# Star Wars Survey
* Does the rest of America realize that “The Empire Strikes Back” is clearly the best of the bunch?
* Is Luke Skywalker really the most popular character?

### Conclusion:
* The original installments : Star Wars 4-6 has better ratings than the new installments Star Wars 1-3; This trend is true across educational background, gender, and region. 
* The most favorite character overall is Luke SkyWalker, however, when we analyze the most favorite character by gender, turns out most female favor Obi Wan Kenobi and Han Solo, while males favor Princess Leia and Luke. 
* Most disliked character by females is Jar Jar Blinks, while by males it's Emperor Palpatine. However, overall, the most disliked character is Emperor Palpatine.

```python
import pandas as pd
#specify an encoding because the data set has some characters 
#that aren't in Python's default utf-8 encoding.
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

## Deleting null Respondent ID

```python
star_wars = star_wars[star_wars['RespondentID'].notnull()]
star_wars.head(3)
```

## Exploring Data

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

## Transforming string data into boolean for analysis

```python
yes_no = {'Yes': True, "No" : False}
x = ['Have you seen any of the 6 films in the Star Wars franchise?','Do you consider yourself to be a fan of the Star Wars film franchise?']
star_wars[x[0]] = star_wars[x[0]].map(yes_no)
star_wars[x[1]] = star_wars[x[1]].map(yes_no)
```

## Transforming string into boolean.
* The title of the Movie (string) indicates the responder has watched the movie, and therefore we assign "True". 
* NaN or null indicates that the responder did not check the checkmark in the survey, we assign "False" for this value.

```python
import numpy as np
i = 3
name = star_wars.iloc[4,3:9]
for title in name:
    colsbool = {}
    colsbool[title] = True
    colsbool[np.NaN] = False
    column_title = star_wars.columns[i]
    star_wars[column_title] = star_wars[column_title].map(colsbool)
    i += 1
```

## Transforming Column Name (Quantity of Watchers)
* How many of the responder watch this movie?
* Renaming these column in the data to a form that is better understood for further analysis.
* From unnamed with number to seen_{movie series number}. For example, "Star Wars I" will be renamed to "seen_1"

```python
i = 1
cols = {}
for seen in star_wars.columns[3:9]:
    cols[seen] = 'seen_{}'.format(i)
    i += 1
star_wars = star_wars.rename(columns=cols)
    
```

## Transforming Column Name (Rating of Movies)
* How will the responder rate this movie?
* Renaming these column in the data to a form that is better understood for further analysis. 
* From unnamed with number to rank_{rating for movie series number}. For example, "rating for Star Wars I" will be renamed to "rank_1"


```python
a = 1
cols2 = {}
for rank in star_wars.columns[9:15]:
    cols2[rank] = 'ranking_{}'.format(a)
    a += 1
star_wars = star_wars.rename(columns=cols2)
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
star_wars.columns[9:15]
```

## Analyzing the Ranking of Each Installment of Movies
* 1 is highest, 2 is high, 3 is moderate, 4 is less, 5 is worst.
* Lower rating is better in this analysis.
* Using plot to visualize the average rating from responder for each movie.

```python
ranking_mean = star_wars[star_wars.columns[9:15]].mean()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

ranking_mean.plot.bar()
```

The highest rating according to audience is the 4th,5th,6th movie, with the champion being Star Wars: Episode V The Empire Strikes Back.
The original installments by far has better rating than the new installments

## Analyzing the Attendance of Each Installment of Movies
* The quantity indicates the popularity.
* Higher rating is better in this analysis.
* Using bar plot to visualize the sum of responder who watch respective movie.

```python
seen = star_wars[star_wars.columns[3:9]].sum()
seen.plot.bar()
```

The result is in proportion with the ratings. The 3rd installments have fewest watchers and lowest rating. The 5th installments have the most watchers and highest rating.

## Analyzing the Ranking and Attendance of Each Installment of Movies by Gender
* for ranking: 1 is highest, 2 is high, 3 is moderate, 4 is less, 5 is worst. Lower rating is better.
* for seen: higher quantities means the movie is more popular. 
* Using bar plot to visualize the comparison of male and females.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
rank_mean_f = females[females.columns[9:15]].mean()
rank_mean_m = males[males.columns[9:15]].mean()
df = pd.DataFrame({'females': rank_mean_f,'males': rank_mean_m})
ax = df.plot.bar(rot=0)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
seen_f = females[females.columns[3:9]].sum()
seen_m = males[males.columns[3:9]].sum()
df2 = pd.DataFrame({'females': seen_f,'males': seen_m})
ax2 = df2.plot.bar(rot=0)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

* The trend is almost the same across gender for ranking. 
* There are more male moviegoers than female ones. 
* The quantity of female moviegoers are higher in the Star Wars I than Star Wars IV.

## Analyzing the Ranking and Attendance of Each Installment of Movies by Education level
 * Using bar plot to visualize the comparison across education levels

```python
star_wars['Education'].value_counts()
```

```python
bool1 = star_wars['Have you seen any of the 6 films in the Star Wars franchise?']==True
star_wars[bool1]['Education'].value_counts()
```

There are only 8.7% of high school graduate have seen the movies from the respondents and 0.3% from the respondents who have less than high school degree. Thus, we will ignore the 0.3% data in this analysis.

```python
def rank_df(srs):
    count = srs.value_counts()
    z = count.index
    edu_rank = {}
    i=0
    for edu in z:
        df1 = star_wars[srs == str(edu)]
        edu_rank[edu] = df1[df1.columns[9:15]].mean()
        i+=1
    df_edur = pd.DataFrame(edu_rank)
    return df_edur

def seen_df(srs):
    count = srs.value_counts()
    z = count.index
    edu_seen = {}
    i=0
    for edu in z:
        df1 = star_wars[srs == str(edu)]
        edu_seen[edu] = df1[df1.columns[3:9]].sum()
        i+=1
    df_edus = pd.DataFrame(edu_seen)
    return df_edus
    
```

```python
df_edur = rank_df(star_wars['Education'])
df_edus = seen_df(star_wars['Education'])
```

```python
axedur = df_edur.plot.bar(rot=0)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
axedus = df_edus.plot.bar(rot=0)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

Most viewers have bachelor degrees, while high school degree attendees are also not that prevalent. On ranking, there is the same trend across education level.

```python
star_wars['Age'].value_counts()
```

```python
df_locr = rank_df(star_wars['Location (Census Region)'])
df_locs = seen_df(star_wars['Location (Census Region)'])
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
axlocr = df_locr.plot.bar(rot=0,ax=axes[0])
axlocr.legend().set_visible(False)
axlocs = df_locs.plot.bar(rot=0,ax=axes[1])

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

East North Central, Pacific, and South Atlantic has the most viewers.

```python
star_wars['Which character shot first?'].value_counts()
df_sr = rank_df(star_wars['Which character shot first?'])
df_ss = seen_df(star_wars['Which character shot first?'])
fig, axes = plt.subplots(nrows=2, ncols=1)
axlocr = df_sr.plot.bar(rot=0,ax=axes[0])
axlocr.legend().set_visible(False)
axlocs = df_ss.plot.bar(rot=0,ax=axes[1])

plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

## Exploring Most Favorite Character
* As we do not have enough information to rename the characters, we will instead match the data with the article where this data is used:
https://fivethirtyeight.com/features/americas-favorite-star-wars-movies-and-least-favorite-characters/


```python
char={}
star_wars = star_wars.rename(columns={star_wars.columns[15]:'Unnamed: 15'})
for srs in star_wars.columns[15:29]:
    char[srs] = star_wars[srs].value_counts()

df_char = pd.DataFrame(char)
df_char = df_char.T
df_char = df_char.sort_values(by=['Very favorably'], ascending = False)
df_char
df_char.plot.barh(figsize=(10,20))
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

Comparing with the articles, we can deduce the unnamed and the name of the character, as the result is closely if not the same with the graph in the article. The only disrepancies might lies in which character is the least favorite, between Jar Jar Blinks or Emperor Palpatine.

```python
cols_char = {'Unnamed: 15': 'Luke Skywalker'
,'Unnamed: 16': 'Yoda'
,'Unnamed: 17': 'R2-D2'
,'Unnamed: 18': 'Darth Vader'
,'Unnamed: 19': 'Princess Leia Organa'
,'Unnamed: 20': 'Jar Jar Binks'
,'Unnamed: 21': 'Anakin Skywalker'
,'Unnamed: 22': 'Padme Amidala'
,'Unnamed: 23': 'Boba Fett'
,'Unnamed: 24': 'C-3P0'
,'Unnamed: 25': 'Obi Wan Kenobi'
,'Unnamed: 26': 'Emperor Palpatine'
,'Unnamed: 27': 'Lando Calrissian'
,'Unnamed: 28': 'Han Solo'}
```

```python
star_wars = star_wars.rename(columns = cols_char)
```

```python
df_char = df_char.T
df_char = df_char.rename(columns = cols_char)
df_char = df_char.T
df_char = df_char.sort_values(by=['Very favorably'], ascending = False)

df_char.plot.barh(figsize=(10,20))
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python
favor = {'Very favorably':5,
'Somewhat favorably':4,
'Neither favorably nor unfavorably (neutral)':3,
'Unfamiliar (N/A)':np.nan,
'Somewhat unfavorably':2,
'Very unfavorably':1}
 
for srs in star_wars.columns[15:29]:
    star_wars[srs] = star_wars[srs].map(favor)
char_rank_mean = star_wars[star_wars.columns[15:29]].mean()
char_rank_mean = char_rank_mean.sort_values(axis=0, ascending=True, inplace=False)
char_rank_mean.plot.bar()
```

## Exploring Most Favorite Character by Gender

```python

def char_fave(df):
    df = df.rename(columns = {df.columns[15]:'Unnamed: 15'})
    df = df.rename(columns = cols_char)
    for srs in df.columns[15:29]:
        df[srs] = df[srs].map(favor)
    char_rank_mean = df[df.columns[15:29]].mean()
    char_rank_mean = char_rank_mean.sort_values(axis=0, ascending=True, inplace=False)
    return char_rank_mean
```

```python
male_char = char_fave(males)

female_char = char_fave(females)

dfchar_gender = pd.DataFrame({'females': female_char,'males': male_char})
dfchar_gender.plot.barh()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

```python

dfchar_gender = dfchar_gender.sort_values(by=['females'], ascending = False)
dfchar_gender.plot.barh()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

#### The result turns out diffrently than overall result. Female respondents favor Obi Wan Kenobi and Han Solo over Luke Skywalker and Princess Leia Organa

```python
dfchar_gender = dfchar_gender.sort_values(by=['males'], ascending = False)
dfchar_gender.plot.barh()
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
```

#### The result turns out diffrently than overall result. Female respondents favor Luke Skywalker and Princess Leia Organa
